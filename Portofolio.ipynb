```python
daily_sales = df.groupby("Date").agg({
    "Total_Revenue": "sum",
    "Jumlah Terjual": "sum"
}).reset_index()
daily_sales
```

```python
sns.lineplot(x="Date", y="Total_Revenue", data=daily_sales)
plt.title("Daily Revenue Over Time")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.lineplot(x="Date", y="Jumlah Terjual", data=daily_sales)
plt.title("Units Sold day by day")
plt.xticks(rotation=45)
plt.show()
```

```python
category_revenue = df.groupby("Category")["Total_Revenue"].sum().sort_values(ascending=False)
category_revenue
```

```python
sns.barplot(x = category_revenue.index, y=category_revenue.values)
plt.title("Category Revenue")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
loc_store = df.groupby(["Store_Location","Payment_Method"])["Total_Revenue"].sum().unstack().fillna(0)
loc_store
```

```python
loc_store.plot(kind = "bar", stacked = True)
plt.title("Revenue by Location and Payment Method")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_products = df.groupby("Product_Name")["Total_Revenue"].sum().sort_values(ascending=False).head(10)
top_products
```
